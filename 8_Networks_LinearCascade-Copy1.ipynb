# Network Motifs: 

Genes are not isolated but part of a network of interaction between many genes. 

- Proteins can act as transcription factors for other genes, that express proteins that are itsef transcription factors. This forms complex gene regulatory networks that ultimately set the identity of a cell.

- When some node of the network changes (for instance, due to a mutation), the behavior of the cell can change (leading to genetic diseases, such as cancer).

- Therefore, to understand and treat genetic diseases, we need  to know the genes involved, how they interact with other genes, and how they ultimately affect cellular responses. 

- The study of full gene regulatory networks is difficult, so a common and modern approach is to look for key network modules, also called __Network Motifs__.

- This minimal __Network Motifs__ are defined as patterns of interconnections occurring in large networks at numbers that are significantly higher than those in randomized networks. 

- The fact that this netorks appear more that what is expecte in random networks is telling is that nature have selected them for a purpose.

We will start our study with the simples way of connecting two genes, which is a linear cascade, i.e., one gene expresses a protein that activates a second "downstream" gene.

## The effect of sequential activation 

We will first study the effect of two sequential genes in a cascade, and compare teh proeries of this simpel network motif with the dynamics with a system of just one single gene. 

This network motif, where things happen as a cascade is very common in signal transduction networks, and seems redundant and irrelevant (what a waste of energy and resources to  do things in two steps instead of one single step). On the other hand, if nature has selected these type of networks with many of thsi linear cascades, it may provide some advantage. To undestand the basics that this strategy introduces, we start by analysing the effect of a sequence of First Order Reactions, such as:

$$X_1\overset{k_1}{\longrightarrow}X_2\overset{k_2}{\longrightarrow}X_3 \tag{38}$$

This system can be solved analytically and compared to a 

$$
A=\begin{bmatrix}
 1 & 0 & 0\\
 0 & 1 & 0\end{bmatrix} ;
B=\begin{bmatrix}
0 & 1 & 0\\
0 & 0 & 1  
\end{bmatrix} ; (B-A)^T= \begin{bmatrix}
-1 & 0\\
 1 & -1 \\
 0 & 1 \end{bmatrix} 
$$

 in this particular case

$$
K=\begin{pmatrix}
 k_1 & 0\\
 0 & k_2
\end{pmatrix}
$$


$$X^A=\begin{pmatrix}
X_1^1 \cdot X_2^0  \cdot X_3^0 \\
X_1^0\cdot X_2^1 \cdot X_3^0
\end{pmatrix} = \begin{pmatrix}
X_1\\
 X_2
\end{pmatrix}
$$

so, using the general formulation of mass action, 

$$
\begin{align}
 \begin{bmatrix}
\frac{\mathrm{d} X_1}{\mathrm{d} t}\\ \frac{\mathrm{d} X_2}{\mathrm{d} t}\\ \frac{\mathrm{d} X_3}{\mathrm{d} t} \end{bmatrix}=(B-A)^T= \begin{bmatrix}
-1 & 0\\
 1 & -1 \\
 0 & 1 \end{bmatrix} \begin{pmatrix}
 k_1 & 0\\
 0 & k_2
\end{pmatrix}\begin{pmatrix}
X_1\\
 X_2
\end{pmatrix}  =  (B-A)^T= \begin{bmatrix}
-1 & 0\\
 1 & -1 \\
 0 & 1 \end{bmatrix}\begin{pmatrix}
k_1 \cdot X_1\\
  k_2 \cdot X_2
\end{pmatrix} =   \begin{bmatrix}
- k_1 \cdot X_1 \\
k_1 \cdot X_1- k_2 \cdot X_2 \\
 k_2 \cdot X_2  \end{bmatrix}
\end{align}
$$


we will use the law of conservation of mass: 



$$ \begin{align*}
C_1 \cdot X_1 + C_2 \cdot X_2 + C_3 \cdot X_3& = cte \\
\end{align*}$$

so
$$ \begin{align*}
C \cdot (B-A)^T &=0\\ 
\begin{pmatrix}C_1 & C_2 & C_3\end{pmatrix} \begin{bmatrix}
-1 & 0\\
 1 & -1 \\
 0 & 1 \end{bmatrix}&=0\\
 -C_1 + C_2 = 0 &\implies  C_1 = C_2\\
 -C_2 + C_3 = 0 &\implies  C_2 = C_3
\end{align*}$$
Therefore,

$$ \begin{align*}
C_1 \cdot X_1(t) + C_1 \cdot X_2(t) + C_1 \cdot X_3(t)&= cte\\
\end{align*}$$

for any value of $C_1$, so taking $C_1=1$, we have 

$$X_1(t)+X_2(t)+X_3(t)=cte $$.

This constant value has to be true also at $t=0$,

$$X_1(t)+X_2(t)+X_3(t)= X_1(0)+X_2(0)+X_3(0) = cte $$.


We will assume, for simplicity, that initially we only have $X_1$ in the system, so the conservation of mass is $$X_1(t)+X_2(t)+X_3(t)=X_1(0)$$

The set of ODEs is:

$$\begin{align}
\frac{dX_1}{dt} &= -k_1 \cdot X_1 \tag{40}\\
\frac{dX_2}{dt} &= k_1 \cdot X_1  - k_2 \cdot X_2 \tag{41}\\
\frac{dX_3}{dt} &= k_2 \cdot X_2 \tag{42}\\
\end{align}$$

From the first case where we solved the first order reaction, we obtain the solution to the first equation, so we rewrite Eq. 15 using this result. We obtain: 

$$ \frac{dX_2}{dt} = k_1 \cdot X_1 (0) \cdot e^{- k_1 t}  - k_2 \cdot X_2 \tag{43}$$

Next, this equation allows us to separate variables, as 

$$ \frac{dX_2}{dt} +  k_2 \cdot X_2 = k_1 \cdot X_1 (0) \cdot e^{- k_1 t}  \tag{44}$$

Again, we need to calculate the integrating factor, $e^{ \int p(x)dx }$, which in this case is $e^{ \int k_2dt }=e^{k_2t}$. We then multiply both terms in eq. 18 by the integrating factor.

$$ \frac{dX_2}{dt}e^{k_2t} +  k_2 \cdot X_2 e^{k_2t}= k_1 \cdot X_1 (0) \cdot e^{- k_1 t}e^{k_2t}=  k_1 \cdot X_1 e^{(k_2-k_1)t}\tag{45}$$

In the next step, we need to recognize that the left-hand side of the equation is the derivative of the product of the dependent variable times the integrating factor:

$$\frac{d}{dt}\left( X_2e^{k_2t}\right)=k_1 X_1 (0) e^{(k_2-k_1)t} \tag{46}$$

We integrate

$$ \int d \left( X_2e^{k_2t}\right)=\int X_1 (0) e^{(k_2-k_1)t}dt \tag{47}$$

$$ X_2e^{k_2t}=\frac{1}{k_2-k_1} k_1 X_1 (0) e^{(k_2-k_1)t}+c \tag{48}$$

$$X_2=\frac{1}{k_2-k_1} k_1 X_1 (0) \frac{e^{(k_2-k_1)t}}{e^{k_2t}}+\frac{c}{e^{k_2t}} \tag{49}$$

$$X_2=\frac{k_1}{k_2-k_1} X_1 (0) e^{-k_1t}+ce^{-k_2t} \tag{50}$$

We have an arbitrary constant because this is a first order differential equation. Let’s calculate $c$  using the initial condition $X_2(0)=0$:

$$0=\frac{k_1}{k_2-k_1} X_1 (0) e^{-k_1 0}+ce^{-k_2 0}=\frac{k_1}{k_2-k_1} X_1 (0) + c \tag{51}$$

$$c=-\frac{k_1}{k_2-k_1} X_1 (0) \tag{52}$$

so finally we obtain the solution for $X_2$:

$$X_2(t)=\frac{k_1}{k_2-k_1}  X_1 (0) e^{-k_1t}-\frac{k_1}{k_2-k_1}  X_1 (0) e^{-k_2t} \tag{53}$$

which reorganizing terms becomes simply:

$$X_2(t)=\frac{k_1 X_1 (0)}{k_2-k_1}  \left(e^{-k_1t}-e^{-k_2t}\right) \tag{54}$$


Now that we have $X_1(t)$ and $X_2(t)$ we can get the expression for $X_3(t)$ by using simply the conservation of mass in Eq. 15

$$X_3(t)=X_1(0)-X_1(t)-X_2(t) \tag{55}$$

plugging the two solutions we obtain

$$X_3(t)=X_1(0)-X_1(0)e^{-k_1t}-\frac{k_1}{k_2-k_1} X_1(0)\left(e^{-k_1t}-e^{-k_2t}\right) \tag{56}$$

that rearranging terms becomes:

$$X_3(t)=X_1(0)\left[1-e^{-k_1t}-\frac{k_1}{k_2-k_1}\left(e^{-k_1t}-e^{-k_2t}\right)\right] \tag{57}$$

```julia
using Plots; gr()
using DifferentialEquations
using ParameterizedFunctions
```

```julia
k_1 = 1.
k_2 = 1.01
X₀ = [0.1,0.0,0.0]
t=collect(0:0.1:10)

plot(t,t->X₀[1]*exp(-k_1*t),label="\\ X_1",seriestype=:line)
plot!(t,t->((k_1*X₀[1])/(k_2-k_1))*(exp(-k_1*t)-exp(-k_2*t)),label="\\ X_2",seriestype=:line)
plot!(t,t->X₀[1] - X₀[1]*exp(-k_1*t) - ((k_1*X₀[1])/(k_2-k_1))*(exp(-k_1*t)-exp(-k_2*t)),label="\\ X_3",seriestype=:line)
title!("Sequential First Order Reaction")
xaxis!("Time")
yaxis!("Concentration")
```

we can compare this with the dynamics of a single reaction occurring in one step

```julia
plot(t,t->X₀[1]*exp(-k_1*t),label="\\ X_1",seriestype=:line)
plot!(t,t->X₀[2]+X₀[1]*(1-exp(-k_1*t)),label="\\ X_2",seriestype=:line)
title!("First order reaction")
xaxis!("Time")
yaxis!("Concentration")
```

As you can see, the fact that we require now the reaction to occur in two sequential steps slows down the reaction and  introduces a delay in the activation of the output. 

### Sequence of casacade of two genes versus one gene 

Now, we will do the same analysis, but in the context of the previously studied Central Dogma approach. For the case of mRNA and protein, we have (which is already a sequence of two events). 

$$
\begin{matrix}
\Psi (T,K,n) &\overset{α_M}{\longrightarrow}& M_1 & \tag{1} \\ 
M_1 &\overset{\gamma_M}{\longrightarrow}& 0 & \\
M_1 &\overset{\alpha_P}{\longrightarrow}& P_1 + M_1 & \\
P_1 &\overset{\gamma_P}{\longrightarrow}& 0 &
\end{matrix}
$$

$$
\begin{matrix}
\Gamma (P_1,K,n) &\overset{α_M}{\longrightarrow} & M_2& \tag{2} \\ 
M_2 &\overset{\gamma_M}{\longrightarrow} & 0 &\\
M_2 &\overset{\alpha_P}{\longrightarrow} & P_2 + M_2&\\
P_2 &\overset{\gamma_P}{\longrightarrow} & 0 &
\end{matrix}
$$

where $\Psi(K,P,n)$ and $\Gamma(P_1,K,n)$ and two Hill functions for activator, and take the form:

$$
\begin{align*}
\Psi (K,P,n) &=  \frac{[T]^n}{K^n+[T]^n} \tag{3}\\
\Gamma (P_1,K,n) &=  \frac{[P_1]^n}{K^n+[P_1]^n} \tag{4}
\end{align*}
$$

Therefore, the set of differential equations that corresponds to this scheme is simply: 

$$
\begin{align*}
   \frac{\mathrm{d} [M_1]}{\mathrm{d} t} &=  -γ_M*M_1+ α_M \frac{T^n}{K^n +T^n}\\
   \frac{\mathrm{d} [P_1]}{\mathrm{d} t} &=   α_P * M_1 - γ_P * P_1 \\
   \frac{\mathrm{d} [M_2]}{\mathrm{d} t} &= -γ_M*M_2+ α_M \frac{P_1^n}{K^n +P_1^n}\\
   \frac{\mathrm{d} [P_2]}{\mathrm{d} t} &=   α_P * M_2 - γ_P * P_2
\end{align*}
$$

```julia
TwoGenes! = @ode_def ab begin
   dM1 = -γ_M*M1+α_M*T^n/(K^n +T^n)
   dP1 =   α_P * M1 - γ_P * P1 
   dM2 = -γ_M*M2+ α_M*P1^n/(K^n +P1^n)
   dP2 =   α_P * M2 - γ_P * P2 
    end α_M γ_M T n α_P γ_P K

OneGene! = @ode_def ab begin
   dM1 = -γ_M*M1+α_M*T^n/(K^n +T^n)
   dP1 =   α_P * M1 - γ_P * P1 
   dM2 = -γ_M*M2+ α_M*T^n/(K^n +T^n)
   dP2 =   α_P * M2 - γ_P * P2 
    end α_M γ_M T n α_P γ_P K
```

When we solve the two systems using the same parameters, we obtain the following curves.

```julia
tspan = (0.0,20.0)
n=1
k_M=0.5
K=1
D=1
α_M=k_M*D
γ_M=1
α_P=1
γ_P2P1=1
γ_P=0.3
T=0.7

p=[α_M,γ_M,T,n,α_P,γ_P,K];
u₀ = [0.01,0.00001,0.01,0.00001]

prob1 = ODEProblem(TwoGenes!,u₀,tspan,p)
sol1 = solve(prob1)
P1=plot(sol1)
title!("Two nodes")
xlabel!("Time [s]")
ylabel!("Concentration [nMol]")

prob2 = ODEProblem(OneGene!,u₀,tspan,p)
sol2 = solve(prob2)
P2=plot(sol2)
title!("One node")
xlabel!("Time [s]")
ylabel!("Concentration [nMol]")


plot(P2,P1, layout=(1,2),legend=true)
```

We see that the addition of a second activator introduces a delay betwene input and output. This is in agreement with the theoretical analysis for a simple linear cascade of interactions. We now from biological systems taht timing is a very relevant issue, specially in developmental biology, since things have to happen in a clear order in time.

### A second activator results in reduced noise transmision from the input to output

We now that gene expression is a noisy process. So, sometimes the cell has to reduce thsi stochasticity to gain some reliability. Let's see what happens if we include some white noise in our simulation. 

```julia
NoiseFunction! = @ode_def negFB begin
   dM1 = 0.3
   dP1 = 0
   dM2 = 0
   dP2 = 0
    end α_M γ_M T n α_P γ_P K
```

```julia
prob6 = SDEProblem(TwoGenes!,NoiseFunction!,u₀,tspan,p)
sol6 = solve(prob6)
P1=plot(sol6,vars=[1,4])
title!("Two genes noise")

prob7 = SDEProblem(OneGene!,NoiseFunction!,u₀,tspan,p)
sol7 = solve(prob7)
P2=plot(sol7,vars=[1,2])
title!("One gene noise")

plot(P2,P1,layout=(1,2),legend=true)
```

we can clearly see that a noisy signal is buffered in a network of two genes, compared to the putput of the network of one gene.

```julia
prob8 = MonteCarloProblem(prob6)
sol8 = solve(prob8,trajectories=100);
#summ8 = MonteCarloSummary(sol8,0:1:20)
#plot(summ8,labels="Middle 95%")
#summ9 = MonteCarloSummary(sol8,0:1:20;quantiles=[0.25,0.75])
summ9 = EnsembleSummary(sol8,0:1:20;quantiles=[0.35,0.65])
plot(summ9,idxs=4,labels="Middle 50%",legend=true)
title!("Response to variability of two genes GRN")
```

```julia
prob8 = MonteCarloProblem(prob7)
sol8 = solve(prob8,trajectories=100);
#summ8 = MonteCarloSummary(sol8,0:1:20)
#plot(summ8,labels="Middle 95%")
summ9 = EnsembleSummary(sol8,0:1:20;quantiles=[0.35,0.65])
plot(summ9,idxs=2,labels="Middle 50%",legend=true)
title!("Response to variability of one gene GRN")
```

## Conclusions

The systems composed of a linear cascade of two genes where the first is an activator of the second some interesting properties compared to a system of one gene with the same kinetic constants:

- It does not affect output intenstity. 
- It introduces delay between input and output (similarly to positive feedback loops with cooperativity). 
- It buffers variations in the intensity of input (similarly to negative feedback loops).
- it buffers variability and noise (similarly to negative feedback loops).

## A linear cascade of repression

Now we will study the same case but in the case of repressors we will compare two cases: one single represor in one step and teh same system followed by second represor in sequential order. 

```julia
TwoGenes_repressor! = @ode_def ab begin
   dM1 = -γ_M*M1+α_M*K^n/(K^n +T^n)
   dP1 =   α_P * M1 - γ_P * P1 
   dM2 = -γ_M*M2+ α_M*K^n/(K^n +P1^n)
   dP2 =   α_P * M2 - γ_P * P2 
    end α_M γ_M T n α_P γ_P K

OneGene_repressor! = @ode_def ab begin
   dM1 = -γ_M*M1+α_M*K^n/(K^n +T^n)
   dP1 =   α_P * M1 - γ_P * P1 
   dM2 = -γ_M*M2+ α_M*K^n/(K^n +T^n)
   dP2 =   α_P * M2 - γ_P * P2 
    end α_M γ_M T n α_P γ_P K
```

```julia
tspan = (0.0,30.0)
n=1
k_M=0.3
K=2
D=1
α_M=k_M*D
γ_M=1
α_P=3
γ_P=0.3
T=1.65

p=[α_M,γ_M,T,n,α_P,γ_P,K];
u₀ = [0.01,0.00001,0.01,0.00001]

prob1 = ODEProblem(TwoGenes_repressor!,u₀,tspan,p)
sol1 = solve(prob1)
P1=plot(sol1,ylims = (0,2.5))
title!("Two nodes")
xlabel!("Time [s]")
ylabel!("Concentration [nMol]")

prob2 = ODEProblem(OneGene_repressor!,u₀,tspan,p)
sol2 = solve(prob2)
P2=plot(sol2,ylims = (0,2.5))
title!("One node")
xlabel!("Time [s]")
ylabel!("Concentration [nMol]")

plot(P2,P1, layout=(1,2),legend=true)
```

We see that, in this case, the system with two repressors is faster than the system with one represor. On the other hand, it introduces a pulse in the amount of output protein. This makes sense becaudse it takes longer to repress the expression, so the dynamcis of expression the protein is faster.

### A linear cascade of two repressors also buffers noise

```julia
prob6 = SDEProblem(TwoGenes_repressor!,NoiseFunction!,u₀,tspan,p)
sol6 = solve(prob6)
P1=plot(sol6,vars=[1,4])
title!("Two genes noise")
xlabel!("Time [s]")
 ylabel!("Concentration [M]")

prob7 = SDEProblem(OneGene_repressor!,NoiseFunction!,u₀,tspan,p)
sol7 = solve(prob7)
P2=plot(sol7,vars=[1,2])
title!("One gene noise")
 xlabel!("Time [s]")
 ylabel!("Concentration [M]")

plot(P2,P1,layout=(1,2),legend=true)
```

```julia
prob8 = MonteCarloProblem(prob6)
sol8 = solve(prob8,trajectories=100);
#summ8 = MonteCarloSummary(sol8,0:1:20)
#plot(summ8,labels="Middle 95%")
summ9 = EnsembleSummary(sol8,0:1:20;quantiles=[0.35,0.65])
plot(summ9,idxs=4,labels="Middle 50%",legend=true)
title!("Response to variability of two repressors")
xlabel!("Time [s]")
 ylabel!("Concentration [M]")
```

```julia
prob8 = MonteCarloProblem(prob7)
sol8 = solve(prob8,trajectories=100);
#summ8 = MonteCarloSummary(sol8,0:1:20)
#plot(summ8,labels="Middle 95%")
summ9 = EnsembleSummary(sol8,0:1:20;quantiles=[0.35,0.65])
plot(summ9,idxs=2,labels="Middle 50%",legend=true)
title!("Response to variability of one repressor")
xlabel!("Time [s]")
 ylabel!("Concentration [M]")
```

## Conclusions

The systems composed of a linear cascade of two genes adds some interesting properties to a signaling cascade:

- It does not affect the ratio input-output 
- It reduces noise.
- A cascade of activators slows the dynamcis and introduces a delay
- A cascade or repression accelerates the dynamics and introduces a pulse
