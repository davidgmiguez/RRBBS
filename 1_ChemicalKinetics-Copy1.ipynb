# Basics of Chemical Dynamics

## 1. Introduction

We will study first how to deal mathematically with systems of interacting entities. Although not limited to chemistry, this is what the field of Chemical Dynamics has been perfecting for many years. Therefore, we will first borrow some ideas from Chemistry that will sound quite familoar to some of the students of this course (but not to others). As a start, let's assume a system of interacting species (chemicals, cells, genes, proteins, organisms...) `A`, `B`, `C` and `D` that interact between them, so the amount of each one is allowed to change over time: 

Some possible schemes of interaction are, for instance:
$$
\begin{align*}
A + B  &\overset{k_1}{\longrightarrow} C  \tag{1} \\
2A + B  &\overset{k_2}{\longrightarrow} C  \tag{2} \\ 
C  &\overset{k_3}{\longrightarrow} A + B   \tag{3} \\ 
AB + C  &\overset{k_4}{\longrightarrow} AC + B   \tag{4} \\ 
AB + CD  &\overset{k_5}{\longrightarrow} AC + BD   \tag{5} \\ 
\end{align*}$$

which is the typycal notation of a chemical reaction, but it can be extrapolated to any other system where some entities interact. We know from Chemistry that the speed of a chemical reaction (i.e., the change in the amount of a given species) is proportional to the amount of collisions (i.e., the amount of times that two given reactants collide into each other in the solution). It is easy to realize that this rate of collisions has to be proportional to the amount of reactants in the solution. For instance, let's illustrate this situation for the the first reaction:

```julia
using Plots 
p1 =scatter(fill(rand(10),1), fill(rand(10),1), fill(rand(10),1), m=(10, 0.8, :blues),framestyle=:box, title="few collisions",label="A")
p1 =scatter!(fill(rand(10),1), fill(rand(10),1), fill(rand(10),1), m=(10, 0.8, :reds),framestyle=:box, title="few collisions",label="B")
p2 =scatter(fill(rand(50),1), fill(rand(50),1), fill(rand(50),1), m=(10, 0.8, :blues),framestyle=:box, title="Many collisions",label="A")
p2 =scatter!(fill(rand(50),1), fill(rand(50),1), fill(rand(50),1), m=(10, 0.8, :reds),framestyle=:box, title="Many collisions",label="B")
plot(p1, p2)
```

It is clear that there will be more collisions in the situation on the right, so the reaction will take place faster. So, not only you will end up with more [C], you will also obtain [C] faster. Based on these basic facts, the speed of each reaction above can be matematically written as:

$$
\begin{align*}
speed~of~reaction~1~&\propto A \cdot B \tag{6} \\ 
speed~of~reaction~2~ &\propto A \cdot A \cdot B \tag{7} \\ 
speed~of~reaction~3 ~ &\propto C \tag{8} \\ 
speed ~of ~reaction~4 ~&\propto AB \cdot B    \tag{8} \\ 
speed ~of ~reaction~5 ~&\propto AB \cdot CD  \tag{10} \\ 
\end{align*}$$

and the proportionality constant is simply the value of the reaction rates $k_i$, which is a measure of how many of the collisions are effective, so:

$$
\begin{align*}
speed~of~reaction~1~&= k_1 \cdot A \cdot B \tag{11} \\ 
speed~of~reaction~2~ &= k_2 \cdot A \cdot A \cdot B \tag{12} \\ 
speed~of~reaction~3 ~ &= k_3 \cdot C \tag{13} \\ 
speed ~of ~reaction~4 ~&= k_4 \cdot AB \cdot B    \tag{14} \\ 
speed ~of ~reaction~5 ~&= k_5 \cdot AB \cdot CD  \tag{15} 
\end{align*}$$

In these system, the equilibrium (defined as the condition of no change in the amount of the species interacting) corresponds to the situation when the speed of the reaction is zero. In these type of __irreversible__ reactions, equilibrium occurs when the reaction is finished because one of the reactants (the limitant) has been fully consumed. On the contrary, in __reversible__ reactions, the equilibrium can be dynamic and does not mean explicitely that the reaction is stopped. For instance, let's asumme a very simple equilibrium between two species:

$$
\begin{align*}
2 NO_2   &\overset{k_1}{\underset{k_2}{\longleftrightarrow}} N_{2}O_4 \tag{16} \\
\end{align*}
$$

This system is actually composed of two different reactions, 

$$
\begin{align*}
2 NO_2   &\overset{k_1}{\longrightarrow}  N_{2}O_4 \tag{17} \\
N_{2}O_4  &\overset{k_2}{\longrightarrow} 2 NO_2 \tag{18}
\end{align*}
$$

and the speed of each reaction is:

$$
\begin{align*}
speed~of~reaction~16~&= k_1 \cdot NO_2 \cdot NO_2 = k_1 \cdot NO_{2}^{2} \tag{19} \\ 
speed~of~reaction~17~ &= k_2 \cdot N_{2}O_4 \tag{20}
\end{align*}$$

For this reversible reaction, it is impossible to have pure $NO_2$ or  $N_{2}O_4$, (i.e., as soon as the amount of one species is approaching zero, the "force" towards the other direction in the reaction approaches infinite) so the equilibrium cannot be defined as the point when no reactions are taking place, or the point where one of the species has been fully consumed. In these conditions, the equilibrium is better defined as the situation when the speed of the two reactions is equal. 
$$
\begin{align*}
k_1 \cdot NO_{2}^{2} = k_2 \cdot N_{2}O_4  \tag{21} 
\end{align*}$$

Thererefore, in conditions of equilibrium in reversible reactions, the ratio $\frac{[NO_2]^2}{[N_{2}O_4]}$ is constant and proportional to the ratios between the reaction rates. Interestingly, this ratio is allways the same. Importantly, this ratio depends on the temperature, because the reactions rates depend on the temperature (i.e., the velocity of the particles depends on the temperature, so number of ''succesfull'' collisions also depens on the temperature). 

A first common characterization of a system of interacting species is the equilibrium constant $K_{eq}$, i.e, the ratio between the two reversible reaction rates. 
$$
\begin{align*}
K_{eq}=\frac{k_2}{k_1} \tag{22} 
\end{align*}$$

This value is very useful because is often tabulated (for chemical reactions) and because it gives you the final ratio of amounts between the reactants. So, for the reaction above, 

$$
\begin{align*}
K_{eq}=\frac{[NO_2]^2}{[N_{2}O_4]} \tag{23} 
\end{align*}$$

With the value of the $K_{eq}$ and the amounts of some of the reactants at equilibrium, we can estimate the value of the others. For instance, let's look at the following reaction: 

$$ H_2+ I_2 \overset{k_1}{\underset{k_2}{\longleftrightarrow}} 2IH \tag{24}$$

At a given temperature, the equilibrium constant is $K_{eq} = \frac{[IH]^2}{[H_2][I_2]} = 0.40$. If [$HI$]=0.75 M and [$H2$]= 0.20 M at equilibrium, then the concentration of [$I_2$] at equilibrium is simply

$$
[I_2] = \frac{[0.75]^2}{[0.2][K_{eq}]} = \frac{[0.75]^2}{[0.2][0.4]} = 7 M \tag{25}
$$

We know that the value of the $K_{eq}$ depends on the temperature, and that the value of $K_{eq}$ sets the balance between the species of the reaction. So, if we plot how the amount of $I_2$ changes with the value of the $K_{eq}$, we obtain the following curve:

```julia
using Plots 
K_eq= LinRange(0.1,1,100)
HI=0.75
H2=0.2
plot(K_eq,(HI^2)./(K_eq.*H2))
title!("Equilibrium concentrations")
ylabel!("I_2 [M]")
xlabel!("Equilibrium constant")
```

which gives us the general result: Changes in the temperature of the system that result in an increase in the $K_{eq}$ shift the equilibrium towards the reactants, while a reduction in $K_{eq}$ shifts the concentrations at equilibrium favoring the products. 

## 2. General Formulation for a system of interacting entities
Let's now introduce a general notation for any interaction scheme between species. Let's define a system where species `A` and `B` react reversibly to give species `C` and `D`:

$$
\begin{align*}
aA + bB  &\overset{k_1}{\underset{k_2}{\longleftrightarrow}} cC + dD \tag{26} \\ 
\end{align*}$$

where `a`, `b`, `c`, `d` correspond to the stoichiometric coefficients for a balanced interaction. At any instant in time, we can define a ratio between the amounts of each species, such as:

$$
Q= \frac{[C]^c[D]^d}{[A]^a[B]^b} \tag{27}
$$
where `Q` is defined in chemistry as the reaction quotient, and measures the relative amounts of the interacting species present during a reaction at a particular point in time. As the time evolves, the system moves towards its equilibrium, and the value of `Q` gradually approaches to the equilibrium constant $K_{eq}$. The general expression of this equilibrium constant is:

$$
K_{eq}= \frac{[C]^c[D]^d}{[A]^a[B]^b} \tag{28}
$$

this apparently peculiar relationship between the amounts of reactants and products in an equilibrium (no matter how many reactants you start with) is based on the previous notion of dynamic equilibrium (two opposite reactions with the same speed). The values of `[A]`, `[B]`, `[C]` and `[D]` represent the amount of each species at equilibirum. If this amount is given as a concentration, lets say in moles/volume, `M`, the units of the equilibrium constant $K_{eq}$ are:

$$
[K_{eq}]=M^{c+d-a-b} \tag{29}
$$

For instance, lets consider the following reaction

$$NaCO_3 + CaCl_2 \overset{k_1}{\underset{k_2}{\longleftrightarrow}} CaCO_3 + 2 \cdot NaCl \tag{30}\\  $$
The equilibrium constant is

$$
K_{eq}= \frac{[CaCO_3][NaCl]^2}{[NaCO_3][CaCl_2]} \tag{31}\\
$$

and the units are

$$
[K_{eq}]=M^{1+2-1-1}=M  \tag{32}\\
$$

Let's say that we can measure the final concentration at equilibrium for each species, for instance:

```julia
NaCO_3=2;
CaCl_2=0.5;
CaCO_3=2;
NaCl=1.2;
```

Then the equilibrium constant can be calculated as:

```julia
K_eq=(CaCO_3*NaCl^2)/(NaCO_3*CaCl_2)
```

## 3. Independence of equilibrium concentrations on initial conditions: 

We have seen how, in reversible reactions, the ratio between the concentrations of the reactants at equilibrium depends only on the ratio between the kinetic constants, and therefore it is __independent on the initial concentrations__ of the reactants. This apparent simple result has many important implications, and therefore, it is important to think a bit about this feature. To illustrate that, we will show the solution of the system for two different initial conditions, and compare the dynamics and the value of ``Q(t)``.

```julia
using DifferentialEquations
using ParameterizedFunctions
```

```julia
simpleODE3! = @ode_def abetterway begin
  da = -k1 * a * b + k2 * c * d^2
  db = -k1 * a * b + k2 * c * d^2
  dc = k1 * a * b - k2 * c * d^2
  dd = 2 * k1 * a * b - 2 * k2 * c * d^2
    end k1 k2
```

```julia
k1=2.3e0;  # units 1/(Ms)
k2=2.5e0;  # units 1/(M M s)
tspan = (0.0,300.0)
p = (k1,k2)
a₀=0.02; # units (M)
b₀=0.01; # units (M)
c₀=0; # units (M)
d₀=0; # units (M)
u₀=[a₀,b₀,c₀,d₀];
prob = ODEProblem(simpleODE3!,u₀,tspan,p)
sol1 = solve(prob);
```

```julia
a₀=0.05; # units (M)
u₀=[a₀,b₀,c₀,d₀];
prob = ODEProblem(simpleODE3!,u₀,tspan,p)
sol2 = solve(prob);
```

```julia
p1=plot(sol1,xlabel="Time [s]",ylabel="Concentration [M]",title="First IC",ylims = (0,0.05));
p2=plot(sol2,xlabel="Time [s]",ylabel="Concentration [M]",title="Second IC",ylims = (0,0.05));
plot(p1,p2,layout=(1,2),legend=true)
```

```julia
Q1=(sol1[3,:].*sol1[4,:].^2)./(sol1[1,:].*sol1[2,:])
Q2=(sol2[3,:].*sol2[4,:].^2)./(sol2[1,:].*sol2[2,:])
plot(sol1.t,Q1,title= "Plot of the quotient coefficient",xlabel="Time [s]",ylabel="Q [M]")
plot!(sol2.t,Q2,title= "Plot of the quotient coefficient",xlabel="Time [s]",ylabel="Q [M]")
```

We see that the value of $Q$ aproaches the same final value at the end of the reaction, and this value is $k_{EQ}$.

## 4. Matrix Formulation for systems of interactions

Sometimes, it is convinient to write down a system of interacting species as a vector-matrix formulation. This will be useful when dealing with large numbers of species and or interactions. Consider the familiar chemical reaction for the formation of water:

$$ 2H_2 + O_2 \overset{k}{\longrightarrow}  2 H_2O \tag{33}$$

The previous equation can be rewritten in a more general form as
 
 $$ 2 X_1 + X_2 \overset{k}{\longrightarrow}  2 X_3 \tag{34}$$

where $X_1$, $X_2$, and $X_3$ are the species $H_2$, $O_2$, and $H_2O$, respectively
 
The first step to write the system as a matrix form, we place all reactants at both sides of the reaction.

  $$ 2 X_1 + X_2  + 0 X_3 \overset{k}{\longrightarrow}  0 X_1 +  0 X_2  + 2 X_3 \tag{35}$$
 
The values in front of each number are the stoichimetric coefficients, which state the number of molecules that participate in each reaction. We can write these stoichiometric coefficiens as
 
$$ A_1 X_1 + A_2 X_2  + A_3 X_3 \overset{k}{\longrightarrow}  B_1 X_1 +  B_2 X_2  + B_3 X_3 \tag{36}$$

$A_1$ = 2, $A_2$ = 1, $A_3$ = 0, $B_1$ = 0, $B_2$ = 0, and $B_3$ = 2.

In compact form, the reaction can be written as:

$$
\sum_{j=1}^{3} A_j X_j \overset{k}{\longrightarrow}  \sum_{j=1}^{3} B_j X_j \tag{37}
$$

or in vector form:$ A X \overset{k}{\longrightarrow} B X  \tag{38}$

that represents the compact general formulation for a system of interacting species. Here, `X` is the state vector (i.e, the vector with all species), and `A` and `B` are the matrices of stoichiometric coefficients. For the particular case of the formation of water, we obtain:

$$\begin{align*}
X&=\begin{bmatrix}X_1\\X_2\\X_3]\end{bmatrix}=\begin{bmatrix}H_2\\O_2\\H_2O\end{bmatrix} \tag{39}\\
A&=\begin{bmatrix}A_1 & A_2 & A_3 \end{bmatrix}=\begin{bmatrix} 2  &1 & 0 \end{bmatrix} \tag{40}\\
B&=\begin{bmatrix} B_1 & B_2 & B_3\end{bmatrix}=\begin{bmatrix} 0 & 0 & 2 \end{bmatrix} \tag{41}\\
\end{align*}$$

To learn how to write reversible reactions as a matrix form, let's practice with the following reaction:

$$NaCO_3 + CaCl_2 \overset{k_1}{\underset{k_2}{\longleftrightarrow}} CaCO_3 + 2 \cdot NaCl \tag{42}\\  $$

We will decouple the reversible reaction into two irreversible steps:

$$\begin{align*}
NaCO_3 + CaCl_2 \overset{k_1}{\longrightarrow} CaCO_3 + 2 \cdot NaCl \tag{43}\\ 
CaCO_3 + 2 \cdot NaCl \overset{k_2}{\longrightarrow} NaCO_3 + CaCl_2 \tag{44}
\end{align*}$$

Using the same matrix notation as in Eq. 13: 

$$\begin{align*}
X_1 + X_2 \overset{k_1}{\longrightarrow} X_3 + 2 \cdot X_4 \tag{45}\\ 
X_3 + 2 \cdot X_4 \overset{k_2}{\longrightarrow} X_1 + X_2 \tag{46}
\end{align*}$$


where $X_1 = Na_2CO_3$, $X_2  = Ca_2Cl$,$X_3  = CaCo_3$, $X_4 = NaCl$. Now we put all reactants at both sides of the reaction.  

$$\begin{align*}
1 \cdot X_1 + 1 \cdot X_2  + 0 \cdot X_3  + 0 \cdot X_4 & \overset{k_1}  \longrightarrow  0 \cdot X_1+ 0 \cdot X_2 + 1 \cdot X_3 + 2\cdot  X_4 \tag{47}\\
 0 \cdot X_1+ 0 \cdot X_2 +1 \cdot X_3 + 2\cdot X_4 &\overset{k_2} \longrightarrow 1 \cdot X_1 + 1\cdot  X_2 + 0 \cdot X_3 + 0 \cdot X_4 \tag{48}
\end{align*}$$

in a more general form:
$$\begin{align*}
A_{11} X_1 + A_{12} X_2  + A_{13} + A_{14} X_4 &\overset{k_1}  \longrightarrow  B_{11}X_1+  B_{12} X_2 +  B_{13} X_3 + B_{14} X_4 \tag{49}\\
A_{21} X_1+ A_{22} X_2 +A_{23} X_3 + A_{24} X_4 & \overset{k_2} \longrightarrow B_{21}  X_1 + B_{22}  X_2 + B_{23} X_3 + B_{24} X_4 \tag{50}
\end{align*}$$

The numbers $A_{ij}$ are the stoichiometric coefficientes of the reactants, $B_{ij}$ are the stoichiometric coefficientes of the products.  Using the matrix notation expalined above, these values are now the components of matrices `A` and `B`:

$$
A=\begin{bmatrix}
 1& 1 & 0 & 0  \\ 
 0 & 0 &1   &2  
\end{bmatrix} ;
B=\begin{bmatrix}
0 & 0 &1   &2 \tag{51}\\ 
1 & 1 &0   &0  
\end{bmatrix}
\\
$$

The numbers of rows $i= 1...r$, corresponds to the numbert of reactions `r` involved. The number of columns $j= 1...s$ corresponds to the number of species involved `s`. In conclusion, when we want to write the mass action law for `r` reactions, `k` is now also a vector with components:

$$k=[k_1,k_2,...k_r]^T \tag{52}$$


altogether we have 


$$
\sum_{j=1}^{s} A_{ij} X_j \overset{k}{\longrightarrow}  \sum_{j=1}^{s} B_{ij} X_j,\,\,\,\, \,\,\,\,\,\,\,\, \,\,\,\,  i=1...r\tag{53}
$$

## 5. Order of a reaction and units of k

The order of a reaction refers to the power dependence of the rate on the concentration of each reactant. In brief, the order of the reaction indicates the correlation of its velocity with the amount of reactants.
- For a ``zero-order reaction``, the rate does not depend on the concentration of any species.
- For a ``first-order reaction``, the rate is dependent on the concentration of a single species.
- For a ``second-order reaction``, the rate is dependent on the square of the concentration of a single reactant, or two reactants.

In Mass-Action kinetics, we can define the reaction order of a given reaction $i =1…r$ as:

$$
\sum_{j=1}^{s} A_{ij} \tag{54}
$$

i.e., the reaction order corresponds to the sum of the stoichiometric coefficients of the species appearing in the reactant of the $i$ reaction. 
For example, in the reaction 

$$NaCO_3 + CaCl_2 \overset{k_1}{\underset{k_2}{\longleftrightarrow}} CaCO_3 + 2 \cdot NaCl \tag{55}\\  $$

with a stoichiometrix coefficient matrix $A$ , 

$$
A=\begin{bmatrix}
 1& 1 & 0 & 0  \\  \tag{56}
 0 & 0 &1  &2  
\end{bmatrix} ; 
$$

The forward reaction $i=1$ has an order or 2, the backward reaction $i=2$ has an order of 3. The units of the kinetic constant $k_i$ of each reaction $i$ can be written in terms of the order of the reaction:

$$[k_i]=\frac{M^{1-\sum_{j=1}^{s} A_{ij}}}{s} \tag{57}$$
